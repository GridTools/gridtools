Let's derive a more convoluted stencil from HorizontalDiffusion:

```python
class MadHD (MultiStageStencil):
    def __init__ (self, domain):
        super ( ).__init__ ( )
        #
        # temporary data fields to share data among the different stages
        #
        self.lap = np.zeros (domain)
        self.fli = np.zeros (domain)
        self.flj = np.zeros (domain)


    def stage_laplace (self, out_lap, in_data):
        for p in self.get_interior_points (out_lap,
                                          ghost_cell=[-1,1,-1,1]):
           out_lap[p] = -4.0 * in_data[p] +  (
                         in_data[p + (-1,0,0)] + in_data[p + (1,0,0)] +
                         in_data[p + (0,-1,0)] + in_data[p + (0,1,0)] )


    def stage_flux_i (self, out_fli, in_data):
        for p in self.get_interior_points (out_fli,
                                           ghost_cell=[-1,0,-1,0]):
            out_fli[p] = in_data[p + (1,0,0)] - in_data[p]


    def stage_flux_j (self, out_flj, in_data):
        for p in self.get_interior_points (out_flj,
                                           ghost_cell=[-1,0,-1,0]):
            out_flj[p] = in_data[p + (0,1,0)] - in_data[p]


    @Stencil.kernel
    def kernel (self, out_data, in_data, in_wgt):
        #
        # Laplace
        #
        self.stage_laplace (out_lap=self.lap,
                           in_data=in_data)

#         for p in self.get_interior_points (self.lap,
#                                            ghost_cell=[-1,1,-1,1]):
#             self.lap[p] = -4.0 * in_data[p] +  (
#                           in_data[p + (-1,0,0)] + in_data[p + (1,0,0)] +
#                           in_data[p + (0,-1,0)] + in_data[p + (0,1,0)] )
        #
        # the fluxes are independent, because they depend on 'self.lap'
        #
        self.stage_flux_i (out_fli = self.fli,
                           in_data  = self.lap)
        self.stage_flux_j (out_flj = self.flj,
                           in_data  = self.lap)

        for p in self.get_interior_points (self.fli,ghost_cell=[-1,0,-1,0]):
           #
           # Trying a=a+b
           #
           self.fli[p] = (self.fli[p + (-1,0,0)] - self.fli[p] +
                          self.flj[p + (0,-1,0)] - self.flj[p] )

        for p in self.get_interior_points (out_data):
            #
            # Last stage
            #
            out_data[p] = in_wgt[p] * (
                          self.fli[p + (-1,0,0)] - self.fli[p] +
                          self.flj[p + (0,-1,0)] - self.flj[p] )
```

#Stage execution experiments

##Indentify input and output data

Let's try to identify stages' input and output data from their data dependency graph

###Laplace stage

This is the dependency graph

```python
lapstg = stencil.stages[0]
lapdd = lapstg.get_data_dependency()
stencil.plot_data_dependency(lapstg.get_data_dependency())
lapdd.nodes()
```

```python
lapstg.outputs = []
lapstg.inputs = []
for node in lapdd.nodes_iter():
    #Output data have no predecessors
    if not lapdd.predecessors(node.name):
        lapstg.outputs.append(node)
    #Input nodes have no successors
    if not lapdd.successors(node.name):
        lapstg.inputs.append(node)
print('Input data:',lapstg.inputs)
print('Output data:',lapstg.outputs)
```

###All stages at once

```python
for stg in stencil.stages:
    print('Stage:', stg.name)
    stg.outputs = []
    stg.inputs = []
    stgdd = stg.get_data_dependency()
    for node in stgdd.nodes_iter():
        #Output data have no predecessors
        if not stgdd.predecessors(node.name):
            stg.outputs.append(node)
        #Input nodes have no successors
        if not stgdd.successors(node.name):
            stg.inputs.append(node)
    print('\tInput data:',stg.inputs)
    print('\tOutput data:',stg.outputs,'\n')
```

Stage 003 (the one assigning to `fli` from `fli` and `flj`) has some problems due to the self looping of `fli`.
While `fli` is both an input and an output, the self loop gives the node both a predecessor and a successor, so it isn't detected as neither input nor output.

```python
stencil.stages[3].get_data_dependency().nodes_with_selfloops()
```

As a temporary solution, let's arbitrarily add the self looping node to stage 003's lists

```python
stg003 = stencil.stages[3]
fli_node = stg003.get_data_dependency().nodes_with_selfloops()[0]
stg003.inputs.append(fli_node)
stg003.outputs.append(fli_node)

print('Input data:',stg003.inputs)
print('Output data:',stg003.outputs,'\n')
```

```python
#Resolve aliases at stencil scope, substituting the alias with the corresponding symbol, that can be found inside the symbol table!
for stg in stencil.stages:
    print('Stage:', stg.name)
    for i, data in enumerate(stg.inputs):
        if data.kind == 'alias':
            stg.inputs[i] = stg.scope.symbol_table[data.value]
    for i, data in enumerate(stg.outputs):
        if data.kind == 'alias':
            stg.outputs[i] = stg.scope.symbol_table[data.value]
    print('\tInput data:',stg.inputs)
    print('\tOutput data:',stg.outputs,'\n')
```

##Build new execution graph

After the StencilInspector has finished its job, the Stencil scope contains a preliminary stage execution graph,
formed by a linear chain of stages in the order they were entered in the stencil kernel.

We can iteratively build the final and correct stage execution graph by traversing the preliminary graph in order:
for each node, we add it in the new graph, and look if any input of the current node matches an output on other nodes of the new graph.

Since all the stages present in the new graph come before or at the same level of the current node (due to the traversal order), we avoid an accidental link with a stage that should depend on the current one.

To avoid double connections, a given input is removed from the available ones after the first match is found from a postorder topological sort of the new execution graph. The sort is executed in postorder to match the latest possible stage between those having the same output data field.

```python
stsc = stencil.scope
new_stage_execution = nx.DiGraph ( )

for stg in stencil.stages:
    new_stage_execution.add_node(stg)
    print('\nProcessing stage',stg.name)
    print('New stage execution nodes:',new_stage_execution.nodes())
    #Create modifiable lists for stage input and output
    stg_in = list(stg.inputs)
    print('Temp inputs',stg_in)
    #Look for connections among nodes introduced in the new graph
    #Use the reverse sort order to find the latest node with the desired output 
    for other in nx.topological_sort(new_stage_execution, reverse=True):
        print('\tLooping others:', other.name)
        #Skip self (obviously...)
        if other is stg:
            continue
        for i, data in enumerate(stg_in):
            ('\t\tInput',i,'-',data)
            if data in other.outputs:
                new_stage_execution.add_edge(other, stg)
                #We have found a match for this input. To avoid double matches, possibly with earlier nodes with the same outputs
                #of "other", pop this stage input from the temporary list
                stg_in.pop(i)
                print('\t\tMatch found! Popping input data. New stg_in:',stg_in)

stencil.plot_data_dependency(new_stage_execution)
```
